This function shows how `plt.legend()` plots the legend using different artist instances and return their names.

#LinePlots

```
import numpy as np

linear_data = np.array([1,2,3,4,5,6])
quadratic_data = linear_data ** 2

#draw the two lines simultaneously 
plt.figure()
plt.plot(linear_data, '-o', quadratic_data, '-*')
```

There are a few things to notice when using `plot` to draw line charts:
- first, here we only give the y-axis value of each data point and matplotlib automatically use the index value as the x-axis values;
- `plot` recognizes that there are two series of data and assigns two different colors to each group. 

##Mini Formatting language of Matplotlib

```
#dash line
plt.plot([12,45,55], '--r') #red dash line
```

```
#add labels and legends
plt.plot([12,45,55], '--r')
plt.plot([1,4,16,60],'-o')
plt.xlabel('Some data')
plt.ylabel('Some other data')
plt.title('A title')
plt.legend(['X-Series', 'Y-Seires'])
```

##The `fill between` function of Matplotlib

Assume that we want to highlight the difference between two curves.

```
plt.plot(linear_data, '--r')
plt.plot(quadratic_data,'-o')
plt.xlabel('Some data')
plt.ylabel('Some other data')
plt.title('A title')
plt.legend(['X-Series', 'Y-Seires'])

#highlight the difference
plt.gca().fill_between(range(len(linear_data)),
                       linear_data, quadratic_data, #lower and upper bond
                       facecolor = 'purple',
                       alpha = 0.25) #transparency
```

##Handling DateTime

The datetime format that NumPy uses is not what Matplotlib expects and we need to reply on the `to_datetime` function of pandas to solve this conflict so we can incorporate NumPy datetime into our charts.

```
import pandas as pd

plt.figure()

observation_dates = np.arange('2020-01-01', '2020-01-09', dtype='datetime64[D]')
observation_dates = map(pd.to_datetime, observation_dates)

plt.plot(observation_dates, linear_data, '-o',
         observation_dates, quadratic_data, '--o')
```

`map` functions returns an iterator that cannot be handled by matplotlib, so we have to convert the mapped result to a list in the first place.

```
import pandas as pd

plt.figure()

observation_dates = np.arange('2020-01-01', '2020-01-07', dtype='datetime64[D]')
observation_dates = list(map(pd.to_datetime, observation_dates)) #to list not generator

plt.plot(observation_dates, linear_data, '-o',
         observation_dates, quadratic_data, '--o')
```

```
plt.figure()

observation_dates = np.arange('2020-01-01', '2020-01-07', dtype='datetime64[D]')
observation_dates = list(map(pd.to_datetime, observation_dates)) #to list

plt.plot(observation_dates, linear_data, '-o',
         observation_dates, quadratic_data, '--o')

rec_gc(plt.gca())
```

##Adjust the axes

```
plt.figure()

observation_dates = np.arange('2020-01-01', '2020-01-07', dtype='datetime64[D]')
observation_dates = list(map(pd.to_datetime, observation_dates)) #to list

plt.plot(observation_dates, linear_data, '-o',
         observation_dates, quadratic_data, '--o')

x = plt.gca().xaxis

#rotate the x-axis
for item in x.get_ticklabels(): #we want to chage the ticklabels of x-axis
  item.set_rotation(45)

#adjust the position of axis
plt.subplots_adjust(bottom=0.25)

#using LaTex in a matplotlib chart
ax = plt.gca()
ax.set_xlabel('Dates')
ax.set_ylabel('Units')
ax.set_title('Quadratic ($x^2$) vs. Linear ($x$) performance')
```

#BarPlots

```
plt.figure()
xvals = np.arange(len(linear_data))

plt.bar(xvals,         #x-axis values
        linear_data,   #height of bars
        width=.35)   
```

```
plt.figure()
xvals = np.arange(len(linear_data))

plt.bar(xvals,         #x-axis values
        linear_data,   #height of bars
        width=.35)  

new_xvals = []
for item in xvals:
  new_xvals.append(item + 0.4)

plt.bar(new_xvals,
        quadratic_data,
        width = 0.35,
        color = 'green')
```

```
#add error bars
from random import randint #not np.random
linear_error = [randint(0,15) for x in range(len(linear_data))]

plt.bar(xvals, linear_data, yerr = linear_error)
```

```
#stacked bar chart
plt.figure()

plt.bar(xvals, linear_data, width=0.3, color='b')
plt.bar(xvals, quadratic_data, width=0.3, bottom=linear_data, color='r')
```

```
#pivot this the above chart to a horizontal one
#width -> height
#bottom -> left
#bar -> barh

plt.figure()

plt.barh(xvals, linear_data, height=0.3, color='b')
plt.barh(xvals, quadratic_data, height=0.3, color='r', left=linear_data)
```

#Dejunkifying a plot

```
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

plt.figure()


languages =['Python', 'SQL', 'Java', 'C++', 'JavaScript']
pos = np.arange(len(languages))
popularity = [56, 39, 34, 34, 29]

#assign the plot to a new variable for further modification
#a container of all the elements
bars = plt.bar(pos, popularity, align='center', 
        linewidth=0, #the edge of each bar
        color='lightslategrey')

#make the bar of Python a contrasting color
bars[0].set_color('#1F77B4')

#set tick and tick labels of x-axis
plt.xticks(pos, languages, alpha=0.8) #location, label, transparency


# remove the Y label since bars are directly labeled
#plt.ylabel('% Popularity',)

# direct label each bar with Y axis values
for bar in bars:
  plt.gca().text(bar.get_x() + bar.get_width()/2, bar.get_height()-5, #location of text, x and y location
                 str(int(bar.get_height())) + '%', #content of the text
                 ha='center', color='w', fontsize='11') #format of the text

#give the plot a title
plt.title('Top 5 Languages for Math & Data \nby % popularity on Stack Overflow', alpha=0.8)

# remove all the ticks (both axes), and tick labels on the Y axis
plt.tick_params(top=False, bottom=False, left=False, right=False, labelleft=False, labelbottom=True)
#bottom, top, left, right,bool Whether to draw the respective ticks.
#labelbottom, labeltop, labelleft, labelright bool Whether to draw the respective tick labels.

# remove the frame of this plot
plt.box(on=False)
#we can also use the following codes to set the spineline of each axis to be invisible
#for spine in plt.gca().spines.values():
#    spine.set_visible(False)
```
